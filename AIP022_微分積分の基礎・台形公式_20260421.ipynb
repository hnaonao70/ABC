## 4．微分
**微分**とは，関数のある点における**変化率**を調べることです．  
授業では，接線の傾きとして説明しました．

$
f'(x)=\lim_{\Delta x\to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x}
$

`sympy.diff()` を使うと，導関数を計算できます．


```python
f = x**3 - 2*x**2 + x + 1
df = sp.diff(f, x)

print("f(x) =", f)
print("f'(x) =", df)

# セル4をログに記録
logger.rec(4)
```

## 5．微分係数と導関数
**微分係数**は，ある1点での接線の傾きです．  
**導関数**は，その傾きを x の関数として表したものです．

例えば，\(f(x)=x^2\) の導関数は \(f'(x)=2x\) です．  
したがって，x=3 における微分係数は 6 になります．


```python
f = x**2
df = sp.diff(f, x)
slope_at_3 = df.subs(x, 3)

print("f(x) =", f)
print("導関数 f'(x) =", df)
print("x=3 における微分係数 =", slope_at_3)

# セル5をログに記録
logger.rec(5)
```

## 7．関数の和や積の微分
微分には重要な公式があります．

- **和の微分**  
$
(f(x)+g(x))' = f'(x)+g'(x)
$

- **積の微分**  
$
(f(x)g(x))' = f'(x)g(x)+f(x)g'(x)
$

Python で実際に確かめてみましょう．


```python
f = x**2 + 1
g = sp.sin(x)

sum_diff = sp.diff(f + g, x)
prod_diff = sp.diff(f * g, x)

print("(f+g)' =", sp.expand(sum_diff))
print("(fg)' =", sp.expand(prod_diff))

# セル7をログに記録
logger.rec(7)
```

## 8．合成関数の微分（連鎖律）
**合成関数**とは，関数の中に別の関数が入っている形です．  
例えば，$f(g(x))$ のような形です．

このときの微分は **連鎖律** を使います．

$
(f(g(x)))' = f'(g(x))g'(x)
$

これは，ニューラルネットワークの**誤差逆伝播**の基本原理でもあります．


```python
g = x ** 2 + 1
f = g ** 3
df = sp.diff(f, x)

print("f(x) =", f)
print("f'(x) =", sp.expand(df))

# セル8をログに記録
logger.rec(8)
```

## 9．勾配降下法
**勾配降下法**は，微分を使って関数の値を小さくしていく方法です．  
授業では，誤差を小さくする学習方法として紹介しました．

$
x_{n+1}=x_n-\eta f'(x_n)
$

ここで，$\eta$ は**学習率**です．


```python
def f(v):
    return v**2 + 2

def df(v):
    return 2*v

eta = 0.1
x_val = 5.0

history = [x_val]
print("更新の履歴 =")
for _ in range(20):
    x_val = x_val - eta * df(x_val)
    print("  ", x_val)
    history.append(x_val)

print("最後の x =", x_val)
print("最後の f(x) =", f(x_val))

# セル9をログに記録
logger.rec(9)
```

## 12．積分・不定積分
**積分**は，微分の逆の考え方を持つ計算です．  
変化を積み重ねて，全体の量を求めるイメージです．

例えば，曲線と x 軸で囲まれた面積を求めるときに使います．

**不定積分**とは，微分すると元の関数になる関数の集まりです．

$
\int f(x)dx = F(x)+C
$

ここで，C は**積分定数**です．

```python
f = 2*x
F = sp.integrate(f, x)

print("f(x) =", f)
print("不定積分 =", F, "+ C")

# セル12をログに記録
logger.rec(12)
```

## 13．定積分
**定積分**は，区間を決めて積分することです．

$

```python
f = x**2
area = sp.integrate(f, (x, 0, 2))

print("∫_0^2 x^2 dx =", area)

# セル13をログに記録
logger.rec(13)
```

## 15．リーマン和
定積分は，小さな長方形の面積を足し合わせる**リーマン和**の極限としても考えられます．

$
\int_a^b f(x)dx \approx \sum_{i=1}^{n} f(x_i)\Delta x
$

区間を細かく分けるほど，より正確になります．


```python
def riemann_sum(func, a, b, n):
    dx = (b - a) / n
    xs = np.linspace(a + dx, b, n)
    return np.sum(func(xs) * dx)

approx = riemann_sum(lambda t: t**2, 0, 2, 1000)
exact = 8 / 3

print("リーマン和による近似 =", approx)
print("正確な値 =", exact)

# セル15をログに記録
logger.rec(15)
```

## 16．台形公式
**台形公式**は，曲線を直線で近似して，台形の面積を足し合わせる方法です．

$
\int_a^b f(x)dx \approx \sum_{i=0}^{n-1}\frac{f(x_i)+f(x_{i+1})}{2}h
$

ここで，$h=(b-a)/n$ です．  
リーマン和より少し精度のよい近似が得られることがあります．

授業では，台形公式を整理すると，**端点は1回，内部の点は2回**現れることを学びました．

$
S=\left(\frac{f(a)+f(b)}{2}+\sum_{i=1}^{n-1}f(a+ih)\right)h
$

この形は，プログラムで実装しやすい形です．


```python
# 台形公式により積分を求める
def trapezoid_rule(func, a, b, n):
    h = (b - a) / n
    S = (func(a) + func(b)) / 2
    for i in range(1, n):
        S += func(a+ i * h)
    return S * h

# 積分したい関数，ここでは f(x)=x^2
def f(x):
    return x**2

# fに台形公式を適用
approx = trapezoid_rule(f, 0, 2, 100)
# 正解
exact = 8 / 3

print("台形公式による近似 =", approx)
print("正確な値 =", exact)
print("誤差 =", abs(approx - exact))

# セル16をログに記録
logger.rec(16)
```
