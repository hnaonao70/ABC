## 7．データを整える
今回は次の2つを行います．

- 「日時」を日付データに変換する

````python
# df_cleanがdfを参照しているため，すべてコピーし，dfから切り離す
df_clean = df_clean.copy()

# 日時を文字列から日付データに変換
df_clean["日時"] = pd.to_datetime(df_clean["日時"])

# 売上金額 = 単価 × 個数
df_clean["売上金額"] = df_clean["単価"] * df_clean["個数"]

# 時間帯を見るために「時」も取り出す
df_clean["時"] = df_clean["日時"].dt.hour

print(df_clean.head())

# セル7をログに記録
logger.rec(7)

# df_clean.head()

df_clean.head()
````

どの商品がよく売れているのかを調べます．  
これは**データサイエンスの基本的な分析**です．

````python
product_summary = df_clean.groupby("商品名")["個数"].sum().sort_values(ascending=False)
print(product_summary)

# セル8をログに記録
logger.rec(8)

product_summary
````

## 9．商品ごとの売上個数をグラフで見る
表だけでは分かりにくいので，棒グラフで見てみます．

````python
# product_summary.plot(kind="bar")
#plt.figure()
plt.bar(product_summary.index, product_summary.values)
plt.title("商品ごとの売上個数")
plt.xlabel("商品名")
plt.ylabel("売れた個数")
plt.xticks(rotation=45)
plt.tight_layout()

# セル9をログに記録
logger.rec(9)

plt.show()
````

## 10．時間帯ごとの売上金額を集計する
今度は，「いつ売れているのか」を見てみます．

````python
hour_summary = df_clean.groupby("時")["売上金額"].sum()
print(hour_summary)

# セル10をログに記録
logger.rec(10)
````

## 11．時間帯ごとの売上金額をグラフで見る
朝，昼，夕方のどこで売上が高いかが分かります．

````python
# hour_summary.plot(kind="bar")
# plt.figure()
plt.bar(hour_summary.index, hour_summary.values)
plt.title("時間帯ごとの売上金額")
plt.xlabel("時")
plt.ylabel("売上金額（円）")
plt.tight_layout()

# セル11をログに記録
logger.rec(11)
plt.show()
````

## 12．分析結果を文章でまとめる
最後に，数字からどんなことが言えそうかを考えます．  
ここが**価値創造**につながる部分です．

````python
top_product = product_summary.index[0]
top_qty = int(product_summary.iloc[0])

top_hour = int(hour_summary.idxmax())
top_sales = int(hour_summary.max())

print(f"最も多く売れた商品は「{top_product}」で，合計 {top_qty} 個でした．")
print(f"最も売上金額が高かった時間帯は {top_hour} 時台で，{top_sales} 円でした．")

# セル12をログに記録
logger.rec(12)
````
