## 5．コサイン類似度
内積は便利ですが，ベクトルの**大きさ**にも影響されます．  
そこで，ベクトルの向きの近さを見るために**コサイン類似度**を使います．

$
\cos \theta = \frac{x \cdot y}{\|x\|\|y\|}
$

値の範囲は **-1 から 1** です．

- 1 に近い：向きがよく似ている  
- 0 に近い：あまり関係がない  
- -1 に近い：逆向き


```python
x1 = np.array([1.0, 0.5, 0.5])
x2 = np.array([1.0, 0.0, 1.0])

dot = np.dot(x1, x2)
norm_x1 = np.linalg.norm(x1)
norm_x2 = np.linalg.norm(x2)
cos_sim = dot / (norm_x1 * norm_x2)

print("x1・x2 =", dot)
print("||x1|| =", norm_x1)
print("||x2|| =", norm_x2)
print("コサイン類似度 =", cos_sim)

# セル5をログに記録
logger.rec(5)
```

## 6. 演習問題 2（コサイン類似度）
次の2つのベクトルのコサイン類似度を求めましょう．

$
u = (1, 2, 0), \quad v = (2, 1, 1)
$

**ヒント**  
- `np.dot(u, v)` で内積  
- `np.linalg.norm(u)` で大きさ


```python
u = np.array([1, 2, 0])
v = np.array([2, 1, 1])

# TODO: コサイン類似度を計算して cos_uv に代入する
norm_u = np.linalg.norm(u)        # LOG
norm_v = np.linalg.norm(v)           # LOG
dot_uv = np.dot(u,v)        # LOG
cos_uv = dot_uv/(norm_u*norm_v)        # LOG

print("コサイン類似度 =", cos_uv)

# セル6をログに記録
logger.rec(6)
```

## 8．行列とベクトルの積
行列 $A$ とベクトル $x$ の積 $Ax$ は，

## 12．転置・単位行列・逆行列
ここでは，重要な**特殊な行列**を確認します．

- **転置行列**：行と列を入れ替えたもの  
- **単位行列**：掛けても元のベクトルや行列を変えないもの  
- **逆行列**：掛けると単位行列になる行列

逆行列は，連立一次方程式を解くときにも使われます．


```python
# 転置が分かりやすい非対称な行列
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("行列 A =")
print(A)

# 転置行列
AT = A.T
print("\n転置行列 A^T =")
print(AT)

# 単位行列
I2 = np.eye(2)
B = np.array([
    [2, 1],
    [3, 4]
])

print("\n2×2の単位行列 I =")
print(I2)

print("\nB =")
print(B)

print("\nI @ B =")
print(I2 @ B)

print("\nB @ I =")
print(B @ I2)

# 逆行列
C = np.array([
    [1, 2],
    [3, 5]
])

C_inv = np.linalg.inv(C)

print("\nC =")
print(C)

print("\nC^{-1} =")
print(C_inv)

print("\nC @ C^{-1} =")
print(C @ C_inv)

print("\nC^{-1} @ C =")
print(C_inv @ C)

# セル12をログに記録
logger.rec(12)
```

## 13．固有値・固有ベクトル
行列をベクトルに掛けると，普通は**向き**も**大きさ**も変わります．  
しかし，向きが変わらず，大きさだけが変わる特別なベクトルがあります．

$
Av = \lambda v
$

このとき，

- $v$：**固有ベクトル**  
- $\lambda$：**固有値**

といいます．

データ分析では，**主成分分析（PCA）** などで重要です．


```python
A = np.array([
    [2, 1],
    [1, 2]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("A =")
print(A)
print("固有値 =")
print(eigenvalues)
print("固有ベクトル =")
print(eigenvectors)

# 1つめの固有値・固有ベクトルで確認
lam = eigenvalues[0]
v = eigenvectors[:, 0]

print("\n確認")
print("v =", v)
print("A @ v =", A @ v)
print("lambda * v =", lam * v)

# セル13をログに記録
logger.rec(13)
```

## 14．連立一次方程式
連立一次方程式は，行列を使って

$
Ax = b
$

の形で書けます．  
つまり，連立方程式の解$x$は，  

$ x = A^{-1}b$   

で求めることができます．


```python
# 連立一次方程式
# 2x + y = 5
# x + 3y = 7

A = np.array([
    [2, 1],
    [1, 3]
])

b = np.array([5, 7])

print("A =")
print(A)

print("\nb =")
print(b)

# 逆行列を求める
A_inv = np.linalg.inv(A)

print("\nA^{-1} =")
print(A_inv)

# 解を計算
x = A_inv @ b

print("\n解 x =")
print(x)

# セル14をログに記録
logger.rec(14)
```
