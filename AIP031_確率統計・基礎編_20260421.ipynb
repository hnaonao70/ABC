## 2. 大数の法則をグラフで見る

### 思い出そう
**大数の法則** とは，試行回数を増やしていくと，標本平均が理論上の平均に近づくという考え方でした．

コイン投げでは，表を1，裏を0とすると，理論上の平均は **0.5** です．  
ここでは，試行回数を増やしながら「表の割合」がどう変化するかをグラフで見ます．

````python
# コイン投げを500回行い，その時点までの表の割合を計算
n = 500
coin = np.random.choice([1, 0], size=n)
running_mean = np.cumsum(coin) / np.arange(1, n + 1)

plt.figure(figsize=(10, 4))
plt.plot(running_mean, label="表の割合")
plt.axhline(0.5, linestyle="--", label="理論値 0.5")
plt.ylim(0,1)
plt.xlabel("試行回数")
plt.ylabel("割合")
plt.title("大数の法則：表の割合は0.5に近づく")
plt.legend()
plt.grid(True)
# plt.show()


# セルをログに記録
logger.rec(4)
````

このグラフでは，最初の方は大きく上下していますが，回数が増えると少しずつ **0.5 に近づく** 様子が見えます．  
この「だんだん安定していく」という感覚は，データ分析でとても重要です．

## 3. 順列と組合せをPythonで確かめる

### 思い出そう
- **順列**：**順番も区別して** 数える
- **組合せ**：**順番を区別せずに** 数える

授業では式で学びましたが，Python では実際に全部並べて確認できます．  
ここでは，数字 1，2，3 から 2つ選ぶ場合を考えます．

````python
items = [1, 2, 3,4,5]

perm = list(permutations(items, 3))
comb = list(combinations(items, 3))

print("順列:", perm)
print("順列の数:", len(perm))
print()
print("組合せ:", comb)
print("組合せの数:", len(comb))


# セルをログに記録
logger.rec(5)
````

### 確認ポイント
例えば `(1, 2)` と `(2, 1)` は，
- **順列** では別もの
- **組合せ** では同じもの

として扱います．

つまり，**順番を見るかどうか** が違いです．  
これは二項分布の理解にもつながる重要な視点です．

## 4. 確率変数と確率分布

### 思い出そう
- **確率変数**：試行の結果を数値で表したもの  
- **確率分布**：それぞれの値がどのくらいの確率で出るかをまとめたもの

ここでは，サイコロの出目を確率変数 \(X\) として，確率分布を表にします．

````python
# サイコロの理論分布
x = np.array([1, 2, 3, 4, 5, 6])
p = np.array([1/6] * 6)

df_dice = pd.DataFrame({
    "出目 X": x,
    "確率 P(X=x)": p
})

print(df_dice)

print("確率の合計:", df_dice["確率 P(X=x)"].sum())

# セルをログに記録
logger.rec(6)
````

授業でも確認したように，**確率分布では確率を全部足すと1になる** 必要があります．

## 6. 確率密度関数と累積分布関数

1.   リスト項目
2.   リスト項目



### 思い出そう
- **確率密度関数（PDF）**：出やすさの濃さ
- **累積分布関数（CDF）**：その値以下になる確率
- 連続分布では，**グラフの高さそのものが確率ではない**
- **面積が確率** になる

ここでは，正規分布の PDF と CDF を描いて比べます．

````python
mu = 170    # LOG （平均μの設定）
sigma =5  # LOG （標準偏差σの設定）

x = np.linspace(mu-4*sigma, mu+4*sigma, 400)
pdf = (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x-mu)**2 /(2*sigma**2))

# 数値積分でCDFを近似
dx = x[1] - x[0]
cdf = np.cumsum(pdf) * dx

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.axhline(0)
plt.title("確率密度関数 PDF")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(mu-10, mu+10)
plt.ylim(0, 0.4)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf)
plt.axhline(0)
plt.title("累積分布関数 CDF")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True)

plt.tight_layout()
# plt.show()

# セルをログに記録
logger.rec(9)
````

左の **PDF** は「どこにデータが集まりやすいか」を表し，右の **CDF** は「左からどれだけ確率がたまったか」を表します．

授業で見たように，ある区間 \([a,b]\) に入る確率は，  
**PDF の面積** で求めることができます．

````python
# 区間 [-1, 1] に入る確率を面積で近似
mask = (x >= -1) & (x <= 1)
prob_between = np.sum(pdf[mask]) * dx

print("P(-1 <= X <= 1) の近似値 =", prob_between)

# セルをログに記録
logger.rec(10)
````

この値は，標準正規分布ではおよそ 0.68 に近くなります．  
つまり，平均の近くに多くのデータが集まることが分かります．

## 7. 期待値を計算する

### 思い出そう
**期待値** は，結果を確率で重みづけした平均でした．  
離散分布では，**値 × 確率** を全部足し合わせて求めます．

まずはサイコロの期待値を計算します．

````python
x = np.array([1, 2, 3, 4, 5, 6])
p = np.array([1/6] * 6)

expected_value = np.sum(x * p)
print("サイコロの期待値 E[X] =", expected_value)

# セルをログに記録
logger.rec(11)
````

サイコロの期待値は **3.5** です．  
3.5 という目は実際には出ませんが，長く繰り返したときの平均的な値を表しています．

### 宝くじの期待値
授業で見たように，期待値は「得か損か」を考えるときにも使えます．  
次のくじを考えます．

- 1万円当たり：1%
- 300円当たり：10%
- はずれ：89%

このとき，もらえる金額の期待値を計算します．

````python
prize = np.array([700_000_000, 150_000_000, 100_000,100_000_000,10_000_000,1_000_000,10_000,3_000,300])
prob = np.array([1/200_000_000, 2 / 200_000_000, 199 / 200_000_000,	1 / 200_000_000,4 / 200_000_000,40 / 200_000_000,600_000 / 200_000_000,2_000_000 / 200_000_000,20_000_000 / 200_000_000
])

expected_prize = np.sum(prize * prob)

print("期待値 =", expected_prize, "円")

# セルをログに記録
logger.rec(12)
````

期待値は **130円** です．  
もし1回300円でこのくじを買うなら，平均すると 130円しか戻ってこないので，長く見ると損になります．  
このように，**期待値は意思決定の判断材料** にもなります．

## 8. 分散と標準偏差

### 思い出そう
- **分散**：平均のまわりにどのくらい散らばっているか
- **標準偏差**：分散の平方根

平均との差をそのまま足すと打ち消し合うため，分散では **差を2乗** して平均します．  
ここでは，同じ平均でもばらつきが違うデータを比べます．

````python
# 2つのデータセット
a = np.array([4, 4.5, 5, 5.5, 6])   # ばらつき小
b = np.array([1, 3, 5, 7, 9])   # ばらつき大

print("Aの平均:", np.mean(a), " 分散:", np.var(a), " 標準偏差:", np.std(a))
print("Bの平均:", np.mean(b), " 分散:", np.var(b), " 標準偏差:", np.std(b))

plt.figure(figsize=(8, 2))
plt.subplot(1, 2, 1)
plt.plot(a, np.zeros_like(a), "o", label="A")
plt.yticks([0], ["A"])
plt.xlabel("値")
plt.xlim(0,10)
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(b, np.ones_like(b), "o", label="B")
plt.yticks([1], ["B"])
plt.xlabel("値")
plt.xlim(0,10)
plt.title("平均は同じでも，ばらつきは違う")
plt.grid(True)
plt.legend()

# セルをログに記録
logger.rec(13)
````

A も B も平均は同じですが，B の方が散らばりが大きいことが分かります．  
この違いを数値として表したものが **分散** や **標準偏差** です．

特に **標準偏差** は元の単位に戻るので，ばらつきを直感的に理解しやすい指標です．
